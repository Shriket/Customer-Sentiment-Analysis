```python
#converting string values in the columns Review Summary & Review detail into list for better analysis. # apply(ast.literal_eval) This function converts each string into a list 
df.loc[:,'Review Summary'] = df['Review Summary'].apply(ast.literal_eval)
df.loc[:,'Review Details'] = df['Review Details'].apply(ast.literal_eval)
```

```python
df = df.explode('Review Summary').reset_index(drop=True) #drop=True - means discard the old index and resetting index(adjusting new index values for exploded column )
# done this only for summary column bcz the review calumn have lots of sentenses and containg words which are not imp for analysis also its quite mixedup sentences & after doing the same for review colm 1543 rows were formed so this df will take more computation thats why ,but already selenium summurized it greatly thats why were using it
#so in this case pandas ,'Review Summary' clm k jitne bhi nye rows bnrhe h usme Review Details me k bhi individual values replicate krrha h jitni rows available h sirf usme hi , baki remaining elements is being excluded
#this index descarding & resetting ((drop=True)) is vimp to get exploded rows in df otherwise therewill be no change in the df(means will not show expploded row kuchbhi kro, kyuki (drop=True) kroge tbhi nyi indexing bnegi)
```

```python

# Exploding the Lists for each product; means transforming each element of a list into a separate row for analysis
df_summary_exploded = df['Review Summary'].explode('Review Summary').reset_index()
df_details_exploded = df['Review Details'].explode('Review Details').reset_index()
```

```python
#dropping the 'Review Summary' & 'Review Details' columns from the original df
df.drop('Review Summary', axis=1, inplace=True)
df.drop('Review Details', axis=1, inplace=True)

#& adding the exploded columns from df_summary_exploded and df_details_exploded to the original df
df["Review Summary"] = df_summary_exploded["Review Summary"]
df["Review Details"] = df_details_exploded["Review Details"]
```

```python

# Removing READ MORE from 'Review Details'
df['Review Details'] = df['Review Details'].str.replace("READ MORE", "", regex=False)
```

NLTK (Natural Language Toolkit)

```python
import nltk
from nltk.corpus import stopwords
```

```python
#Downloading used nltk packages
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('words')
nltk.download('stopwords')
```

```python
example = df['Review Details'][3]
example
```

```python
#Smartly Splits the text into parts so that the model can interpret it
tokens = nltk.word_tokenize(example)
tokens
```

```python
#Part of speech tagging. Determines adjectives, nouns etc.
tagged = nltk.pos_tag(tokens)
tagged
```

```python
#Remove Stopwords such as 'The' 'is' 'and', etc.
stop_words = set(stopwords.words('english')) #stopwords.words('english'): This function gets the set of stop words in English lang
filtered_words = [word for word in tokens if word.lower() not in stop_words] 
#its a list comprehension to create a new list that contains only the words from tokens that are not in stop_words. & word.lower()- ensures the comparison is case-insensitive
filtered_words
```
